```python
import matplotlib.pyplot as plt
```

```python
days = [10,20,30,40,50,60,70]
minimum = [4,4.3,4.2,4.1,3.7,3.5,3.6]
temperature = [4.5,4.8,4.8,4.6,4.0,4.2,4.1]
maximum = [5.0,5.3,5.2,5.1,4.6,4.7,4.5]
```

```python
plt.grid()
plt.title("Weather Graph")
plt.xlabel('days')
plt.ylabel('temperature')
plt.ylim([3,6])

plt.plot(days,temperature,label="Average")
plt.plot(days,minimum,label="Minimum")
plt.plot(days,maximum,label="Maximum")
plt.legend(loc='best')
plt.show()
```
