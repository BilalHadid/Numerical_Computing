```python
import numpy as np
import matplotlib.pyplot as plt


print("\tSolving the temperature with in the slab")

print("------------------------------------------")

print("")



axis_size = 120     

side_length = 10  

dx = side_length / axis_size 

axis_points = np.linspace( 0 , 10 , axis_size )   

colourMap = plt.cm.jet


print("")






T = 5

T = float( T )                            

k = 1.011                                 

dt = ( ( 1 / axis_size ) ** 2 ) / ( 2 * k )

n = int ( T / dt )                              



print("")

print("Please Enter the Maximum Temperature for Middle Point (in degrees): ")

max_temp = input( )

max_temp = float ( max_temp )



def temp_init ( x , y , c , m_temp ) :

    return m_temp * np . exp ( - ( ( x - side_length / 2 ) ** 2 + ( y - side_length / 2 ) ** 2 ) / c ** 2 )


X , Y = np . meshgrid ( axis_points , axis_points )


c = 2


U = temp_init ( X , Y , c , max_temp )


print("")


print("Please Enter the temperature for Left Edge (in degrees): ")

U [ : , 0 ] = input ()

print("")

print("Please Enter the temperature for Right Edge (in degrees): ")

U [ : , -1 ] = input ()

print("")

print()

U [ 0 , : ] = input ("Please Enter the temperature for Top  (in degrees): ")

print("")

print("Please Enter the temperature for Bottom Edge (in degrees): ")

U [ -1 , : ] = input ()

if (max_temp < U.any() ) : 
    print("Error !!!")
    print("Maximum Temperature is lesser than one of the edge's temperature !!!")

B1 = U [ : , 0 ]

B2 = U [ : , -1 ]

B3 = U [ 0 , : ] 

B4 = U [ -1 , : ]

colourMap = plt.cm.jet
plt.imshow (U , cmap =colourMap, vmin = 0 , vmax = max_temp , extent = [ 0 , 20 , 0 , 20 ] )

plt.tick_params (axis = 'both' , which = 'major' , labelsize = 16 )

plt.xlabel('Horizontal (cm)' , fontsize = 16 )

plt.ylabel('Vertical (cm)', fontsize = 16 )

plt.title( 'center point Temperature' )

plt.show( )






def laplacian ( Z , d_ex ) :

    Ztop = Z [ 0 : -2 , 1 : -1 ]

    Zleft = Z [ 1 : -1 , 0 : -2 ]

    Zbottom = Z [ 2 : , 1 : -1 ]

    Zright = Z [ 1 : -1 , 2 : ]

    Zcenter = Z [ 1 : -1 , 1 : -1 ]

    return (Ztop + Zleft + Zbottom + Zright - 4 * Zcenter) / d_ex ** 2



for i in range( n ) :

    deltaU = laplacian ( U , dx )

    Uc = U [ 1 : -1 , 1 : -1 ]

    U [ 1 : -1 , 1 : -1 ] = Uc + dt * ( k * deltaU )

 
    U [ : , 0 ] = B1

    U [ : , -1 ] = B2

    U [ 0 , : ] = B3

    U [ -1 , : ] = B4


plt.imshow ( U , cmap = colourMap, vmin = 0 , vmax = max_temp , extent = [ 0 , 10 , 0 , 10 ] )

plt.tick_params ( axis = 'both' , which = 'major' , labelsize = 16 )

plt.xlabel('Horizontal  (cm)' , fontsize = 16 )

plt.ylabel('Vertical  (cm)' , fontsize = 16 )

plt.title('Final Temperature Result')

plt.show( )
```
